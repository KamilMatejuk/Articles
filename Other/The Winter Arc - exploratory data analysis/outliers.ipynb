```python
import seaborn as sns
import matplotlib.pyplot as plt

import const
from read import read_data
from graphs import set_style, C_DEFAULT
```

```python
data = read_data()
data.head()
```

### Visualize data ranges and outliers

```python
cols = const.COLS_EATING + const.COLS_SLEEPING + const.COLS_READING + const.COLS_WORKING + const.COLS_LEISURE + const.COLS_EXERCISE[-1:]

fig, axes = plt.subplots(len(cols), 1, figsize=(10, 0.5 * len(cols) + 1))
for ax, col in zip(axes, cols):
    sns.boxplot(x=data[col], color=C_DEFAULT, ax=ax)
    set_style(ax, data[col])
    ax.set_ylabel(col.replace('_', '\n'))
    ax.set_xlabel('')
fig.tight_layout()
plt.show()
```
