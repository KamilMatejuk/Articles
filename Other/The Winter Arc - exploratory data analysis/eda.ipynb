## Columns and types

```python
data = pd.read_csv('data.csv')

data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')

for col, typ in [
    ('week_day', 'category'),
    ('working', 'bool'),
    ('out_of_home', 'bool'),
    ('energy', 'category'),
    *[(c, 'float') for c in const.COLS_SLEEPING],
    *[(c, 'float') for c in const.COLS_READING],
    *[(c, 'float') for c in const.COLS_WORKING],
    *[(c, 'float') for c in const.COLS_LEISURE],
    *[(c, 'float') for c in const.COLS_EXERCISE],
    *[(c, 'float') for c in const.COLS_EATING],
    *[(c, 'bool') for c in const.COLS_PUBLISHING],
]: data[col] = data[col].astype(typ)

COLS_BOOLEAN = data.select_dtypes(include=['bool']).columns
COLS_NUMERICAL = data.select_dtypes(include=['float', 'int']).columns
COLS_CATEGORICAL = data.select_dtypes(include=['category']).columns

COLS_TIME = [c for c in data.columns if c.endswith('_time')]

data.tail()
```
