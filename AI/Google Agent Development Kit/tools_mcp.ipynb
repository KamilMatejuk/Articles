```python
from google.genai import types
from google.adk.agents import Agent
from mcp import StdioServerParameters
from google.adk.runners import InMemoryRunner
from google.adk.models.google_llm import Gemini
from google.adk.tools.mcp_tool.mcp_toolset import McpToolset
from google.adk.tools.mcp_tool.mcp_session_manager import StdioConnectionParams
```

```python
retry_config=types.HttpRetryOptions(attempts=3, exp_base=5, initial_delay=1, http_status_codes=[429, 500, 503, 504])
model = Gemini(model="gemini-2.5-flash-lite", retry_options=retry_config)
```

```python
mcp_sum_server = McpToolset(
    connection_params=StdioConnectionParams(
        server_params=StdioServerParameters(
            command="npx", args=["-y", "@modelcontextprotocol/server-everything"],
            tool_filter=["get-sum"],
        )
    )
)
```

```python
math_teacher_agent = Agent(
    name="MathTeacherAgent",
    model=model,
    instruction="""
    You are a math teacher.
    Help me solve math problems.
    Use get-sum tool from mcp_sum_server to make calculations.
    """,
    tools=[mcp_sum_server],
)
```

```python
runner = InMemoryRunner(agent=math_teacher_agent)
_ = await runner.run_debug("What is 2 + 2?")
```
