```python
from google.genai import types
from google.adk.agents import Agent
from google.adk.runners import Runner
from google.adk.models.google_llm import Gemini
from google.adk.sessions import InMemorySessionService
from google.adk.apps.app import App, EventsCompactionConfig
```

```python
retry_config=types.HttpRetryOptions(attempts=3, exp_base=5, initial_delay=1, http_status_codes=[429, 500, 503, 504])
model = Gemini(model="gemini-2.5-flash-lite", retry_options=retry_config)
```

```python
assistant_agent = Agent(
    name="AssistantAgent",
    model=model,
    instruction="You are a personal assistant",
)
```

```python
session_service = InMemorySessionService()
```

```python
assistant_app = App(
    name="AssistantApp",
    root_agent=assistant_agent,
    events_compaction_config=EventsCompactionConfig(compaction_interval=3, overlap_size=1),
)
```

```python
runner = Runner(app=assistant_app, session_service=session_service)
```

```python
_ = await runner.run_debug("I love eating ice cream. What are some other desserts I might like?")
```

```python
_ = await runner.run_debug("What do you know about my dessert preferences?")
```

```python
_ = await runner.run_debug("I also love cookies.")
```
