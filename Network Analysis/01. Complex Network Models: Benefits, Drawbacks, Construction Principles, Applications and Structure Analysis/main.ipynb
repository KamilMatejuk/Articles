# Random Network

```python
GRAPHS = [
    nx.erdos_renyi_graph(n=25, p=0.3),
    nx.erdos_renyi_graph(n=50, p=0.2),
    nx.erdos_renyi_graph(n=250, p=0.1),
    nx.erdos_renyi_graph(n=1000, p=0.01),
]
LABELS = [
    'G(n=25, p=0.3)',
    'G(n=100, p=0.2)',
    'G(n=250, p=0.1)',
    'G(n=1000, p=0.01)',
]
```
